```python
import pandas as pd
df = pd.read_csv('DataSetk-means.csv')
df = df.drop(['rain_1h', 'snow_1h'],axis = 1)
df.head()
```

```python
dfn = df[['temp','clouds_all' , 'traffic_volume']]
from sklearn import preprocessing

x = dfn.values 
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
dfn = pd.DataFrame(x_scaled)
dfn.head()
```

```python
df  = pd.concat([df, dfn], axis=1)
df.head()
```

```python
df = df.drop(['temp','clouds_all' , 'traffic_volume'],axis = 1)
df.head()
```

```python
import random
import math
import statistics
def kmeans(K , data):
    centroids = []
    lists = []
    for  x in range(K):
        ind = random.randint(0,data.shape[0])
        centroids.append([data.iloc[ind,0] ,data.iloc[ind,1] , data.iloc[ind,2] , data.iloc[ind,3] , data.iloc[ind,4]])
        lists.append([])
    for y in range(50): # 50 is no of iteration 
        lists= []
        for  x in range(K):
            lists.append([])    
        for i in range(data.shape[0]):
            dists = []
            for c in centroids:
                dis = []
                if data.iloc[i,0] == c[0]:
                    dis.append(0)
                else:
                    dis.append(1)
                if data.iloc[i,1] == c[1]:
                    dis.append(0)
                else:
                    dis.append(1)
                dis.append(data.iloc[i,2] - c[2])
                dis.append(data.iloc[i,3] - c[3])
                dis.append(data.iloc[i,4] - c[4])
                dis = [j*j for j in dis]
                su = 0
                for x in dis:
                    su+=x
                dists.append(  math.sqrt(su) )
            mi = dists[0]
            minc = 0
            for j in range(1,K):
                if dists[j] < mi:
                    mi = dists[j]
                    minc = j
            lists[minc].append(i)

        for i in range(K):
            l0 = []
            l1 = []
            l2 = []
            l3 = []
            l4 = []
            for ind in lists[i]:
                l0.append(df.iloc[ind,0])
                l1.append(df.iloc[ind,1])
                l2.append(df.iloc[ind,2])
                l3.append(df.iloc[ind,3])
                l4.append(df.iloc[ind,4])
            centroids[i][0] = statistics.mode(l0)
            centroids[i][1] = statistics.mode(l1)
            centroids[i][2] = statistics.mean(l2)
            centroids[i][3] = statistics.mean(l3)
            centroids[i][4] = statistics.mean(l4)
    return lists,centroids

ans,centroids = kmeans(7,df)
# here ans is cluster and centroids is centroid of cluster 
```

```python
lo = []
for ind in ans[5]:
    lo.append(df.iloc[ind,1])
statistics.mode(lo)
# lo is class of cluster 
```

```python
import scipy.io
mat = scipy.io.loadmat('2010rainfall.mat')
mat1 = scipy.io.loadmat('2011rainfall.mat')
```

```python
import numpy as np
np.shape(mat['XR1'])
```

```python
mat1
```

```python
y = mat['ZR1']
y1 = mat1['ZR']
```

```python
x = mat['XR1']
x1 = mat1['XR']
```

```python
mat1
```

```python
pdf1 = pd.DataFrame(x.T)
pdf2 = pd.DataFrame(x1.T)

```

```python
pdf1['class'] = y.T
pdf2['class'] = y1.T
```

```python
np.shape(y)
```

```python
pdf3 = pdf1.append(pdf2)
pdf3.reset_index(drop =True,inplace =True)
```

```python
pdf1
```

```python
y11 = pdf1['class']
X11 = pdf1.drop(['class'], axis=1)
y12 = pdf2['class']
X12 = pdf2.drop(['class'], axis=1)
```

```python
import warnings
warnings.filterwarnings("ignore")
#for s = 42
X11['T-1'] = 0
X11['T-1'][0] = 0
X11['T-1'][1:] = X11[41][:121]
X11['T-2'] = 0
X11['T-2'][0] = 0
X11['T-2'][1] = 0
X11['T-2'][2:] = X11[41][:120]
Xtrain = X11[['T-1','T-2']]
ytrain = X11[41]
X12['T-1'] = 0
X12['T-1'][0] = 0
X12['T-1'][1:] = X11[41][:121]
X12['T-2'] = 0
X12['T-2'][0] = 0
X12['T-2'][1] = 0
X12['T-2'][2:] = X11[41][:120]
Xtest = X12[['T-1','T-2']]
ytest = X12[41]
from sklearn import datasets, linear_model, metrics
reg = linear_model.LinearRegression() 
reg.fit(Xtrain, ytrain) 
y_pred = reg.predict(Xtest)

print('Mean Absolute Error:', metrics.mean_absolute_error(ytest, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(ytest, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(ytest, y_pred)))

print('Coefficients: \n', reg.coef_)
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(Xtrain,ytrain)
y_pred = lasso.predict(Xtest)
print('Coefficients: \n', lasso.coef_) 

sorted(y_pred,reverse=True)[:5]
```
